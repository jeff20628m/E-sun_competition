```python
import os
import sys
import shutil
import glob
ori_fold = '../../../data_set//train' ## orignal_dataset_path
gen_fold ='../../../data_set//format_train/' ## new_dataset_path

def gen_folder(im_class,img_path):
    gen = os.path.join(gen_fold,im_class)
    try:
        os.makedirs(gen)
    except:
        pass
    ## copy file
    shutil.copyfile(img_path, os.path.join(gen,img_path.split('\\')[-1]))

all_file = glob.glob(ori_fold + '/*.jpg')
file_len = len(all_file)
for num,i in enumerate(all_file):
    img_name = i.split('\\')[-1][:-4] ## get img name "e.g XXX_CLASS.jpg" and replace ".jpg" --> "XXX_CLASS"
    img_class = img_name.split('_')[-1] ## get img_class "CLASS"
    gen_folder(img_class,i)
    if num+1 % 100 == 0:
        print("{}/{}".format(num+1,file_len)) ## print out process
```
